# Линейная регрессия

#### Задание
* Реализовать алгоритм линейной регрессии на языке Python
* Спецификация:
    1. Функция f(x) = 7*x
    2. Алгоритм автоматически генерирует исходный набор данных,используя псевдослучайные числа. Размер набора не менее 1000 элементов
    3. Алгоритм позволяет обучить модель и доказать ее точность
    4. Алгоритм визуализирует полученный результат

```python
import numpy as np
import matplotlib.pyplot as plt


eps = 0.0000001
weight0 = 0.0
weight1 = 0.0
new_weight0 = 0.0
new_weight1 = 0.0
t = 0.0
init_step = 0.00005

def difQ_w0(w0, w1, x, y):
    differential = 0
    for i in range(len(x)):
        differential += w1 * x[i] + w0 - y[i]
    return 2 / len(x) * differential

def difQ_w1(w0, w1, x, y):
    differential = 0
    for i in range(len(x)):
        differential += (w1 * x[i] + w0 - y[i]) * x[i]
    return 2 / len(x) * differential

size = 1000

x = np.random.uniform(0, 500, size)
y = [7.0 * x[i] for i in range(size)]

while True:
    t += 1.0
    step = init_step / t
    new_weight0 = weight0 - step * difQ_w0(weight0, weight1, x, y)
    new_weight1 = weight1 - step * difQ_w1(weight0, weight1, x, y)

    if(((new_weight0 - weight0) ** 2 + (new_weight1 - weight1) ** 2) ** 0.5) < eps:
        break
        
    weight0 = new_weight0
    weight1 = new_weight1
    
print(weight0, weight1)
```

### График исходной функции

```python
plt.plot(x, y, 'r-')
plt.grid(True)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

### График функции, полученной в результате обучения

```python
y_new = [weight1 * x[i] + weight0 for i in range(size)]

plt.plot(x, y_new)
plt.grid(True)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
